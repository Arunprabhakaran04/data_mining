```python
import pandas as pd
import numpy as np
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def load_documents(file_path):
    with open(file_path, "r", encoding = "utf-8") as file:
        return file.read().lower()
    
doc1 = load_documents("Document1.txt")
doc2 = load_documents("Document2.txt")
query = load_documents("Query_Document.txt")        
```

```python
#preprocessing with TF-IDF 
vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform([doc1,doc2, query])
```

```python
#compute cosine similarity - 
similarity_matrix = cosine_similarity(tfidf_matrix)
```

```python
querydoc1 = similarity_matrix[2, 0]
querydoc2 = similarity_matrix[2, 1]

print(querydoc1)
print(querydoc2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (6,4))
sns.heatmap(similarity_matrix, annot = True, cmap = "Blues", xticklabels=["doc1", "doc2", "Query"], yticklabels=["doc1", "doc2", "query"])
```
